```python
from IPython.display import clear_output
import random

# this function prints grid and all instances of objects
def makeGrid(rows, cols, kangaroos, player, egg, door):
    print("Don't get eaten by the Kangaroo!")
    print("Each egg you eat will allow you to advance to the next level, simply get to the door!")
    for y in range(rows):
        print(' ---'*cols)
        for x in range(cols):
            kangaroo_created = False
            for kangaroo in kangaroos:
                if kangaroo.coords == [x, y] and x == cols - 1 and kangaroo_created == False:
                    print('| k |', end='')
                    kangaroo_created = True
                elif kangaroo.coords == [x, y] and kangaroo_created == False:
                    print('| k ', end='')
                    kangaroo_created = True
            if player.coords == [x, y] and x == cols - 1 and kangaroo_created == False:
                print('| p |', end='')
            elif player.coords == [x, y] and kangaroo_created == False:
                print('| p ', end='')
            elif egg.coords == [x, y] and x == cols - 1 and kangaroo_created == False:
                print('| e |', end='')
            elif egg.coords == [x, y] and kangaroo_created == False:
                print('| e ', end='')
            elif door.coords == [x, y] and x == cols - 1 and kangaroo_created == False:
                print('| d |', end='')
            elif door.coords == [x, y] and kangaroo_created == False:
                print('| d ', end='')
            elif x == cols - 1 and kangaroo_created == False:
                print('|   |', end='')
            elif kangaroo_created == False:
                print('|   ', end='')
        print()
        if y == rows - 1:
            print(' ---'*cols)

class Kangaroo():
    def __init__(self, coords):
        self.coords = coords

    def initCoords(self, rows, cols):
        self.coords = [random.randint(0, cols-1), random.randint(0, rows-1)]

class Player():
    def __init__(self, coords, egg_created=0):
        self.coords = coords
        self.egg_created = egg_created

    def initCoords(self, rows, cols, kangaroos):
        # initialize first coords
        self.coords = [random.randint(0, cols-1), random.randint(0, rows-1)]
        # check and loop until coords are not same as kangaroo
        for kangaroo in kangaroos:
            while kangaroo.coords == self.coords:
                self.coords = [random.randint(0, cols-1), random.randint(0, rows-1)]

    def movePlayer(self):
        ans = input('Where would you like to move? ')
        if ans == 'up':
            self.coords = [self.coords[0], self.coords[1] - 1]
        elif ans == 'down':
            self.coords = [self.coords[0], self.coords[1] + 1]
        elif ans == 'left':
            self.coords = [self.coords[0] - 1, self.coords[1]]
        elif ans == 'right':
            self.coords = [self.coords[0] + 1, self.coords[1]]

    def checkEgg(self, egg):
        if self.coords == egg.coords:
            self.egg_created += 1
            egg.coords = [-1, -1]
            print('Way to go! Eggs: {} '.format(self.egg_created)) # displays a message when user gets egg (egg)
            # this line doesn't run

class Egg():
    def __init__(self, coords):
        self.coords = coords

    def initCoords(self, rows, cols):
        self.coords = [random.randint(0, cols-1), random.randint(0, rows-1)]

class Door():
    def __init__(self, coords):
        self.coords = coords

    def initCoords(self, rows, cols):
        self.coords = [random.randint(0, cols-1), random.randint(0, rows-1)]

# create a game over function
def game_over(player, kangaroos, door, num_eggs):
    for kangaroo in kangaroos:
        if player.coords == kangaroo.coords:
            return 1
    if player.coords == door.coords and num_eggs == player.egg_created:
        return 2
    return False

# initialize game over flag
flag = False

level = 0

# main outer initializing loop
while True:
    # size of grid
    rows = 5
    cols = 5
    level += 1

    # objects for game
    kangaroos = [Kangaroo([0, 0]) for i in range(level)]
    player = Player([1, 1])
    egg = Egg([2, 2])
    door = Door([3, 3])

    # randomly initiating coordinates for start
    for kangaroo in kangaroos:
        kangaroo.initCoords(rows, cols)
    player.initCoords(rows, cols, kangaroos)
    egg.initCoords(rows, cols)
    door.initCoords(rows, cols)


    # main game loop
    while True:
        clear_output()
        # show grid
        makeGrid(rows, cols, kangaroos, player, egg, door)
        # move player
        player.checkEgg(egg)
        player.movePlayer()
        

        # check game_over
        flag = game_over(player, kangaroos, door, 1)

        if flag == 1:
            print('You were eaten by the kangaroo!')
            level -= 1
            break
        elif flag == 2:
            print('Congrats you beat this level!')
            break

    ans = input('Would you like to play again? ')
    if ans == 'no':
        break
    else:
        continue
```
